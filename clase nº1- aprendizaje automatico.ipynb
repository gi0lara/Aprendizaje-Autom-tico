# Ejercicios de Arrays y Matrices con NumPy

#### Ejercicio 1: Crear un array de 10 elementos con ceros.

```python
import numpy as np

array_cero = np.zeros(10)
print(array_cero)
```

#### Ejercicio 2: Genera un array con los números del 1 al 20.

```python
array1_20 = np.arange(1, 21)
print(array1_20 )
```

#### Ejercicio 3: Crea un array con los números pares del 2 al 50.

```python
array_par = np.arange(2, 51, 2)
print(array_par)
```

#### Ejercicio 5: Crea una matriz de 4x4 con valores aleatorios entre 0 y 1.

```python
matriz = np.random.rand(4, 4)
print(matriz)
```

#### Ejercicio 6: Realiza la suma elemento a elemento de los arrays a=[1,2,3], b=[4,5,6].

```python
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])

suma = a + b
print(suma)
```

#### Ejercicio 7: Multiplica dos matrices 3x3 de enteros aleatorios entre 1 y 10.

```python
matriz_a = np.random.randint(1, 11, (3, 3))
matriz_b = np.random.randint(1, 11, (3, 3))

resultado = np.dot(matriz_a, matriz_b)

print("Matriz A:\n", matriz_a)
print("Matriz B:\n", matriz_b)
print("Resultado:\n", resultado)
```

#### Ejercicio 8: Crea un array de 15 valores aleatorios entre 1 y 100 e imprime su máximo, mínimo y promedio.

```python
array = np.random.randint(1, 101, 15)

print("Array:", array)
print("Máximo:", array.max())
print("Mínimo:", array.min())
print("Promedio:", array.mean())
```

#### Ejercicio 9: Genera una matriz 5x5 con valores del 1 al 25 y extrae su diagonal.

```python
matriz = np.arange(1, 26).reshape(5, 5)

diagonal = np.diag(matriz)

print(matriz)
print("Diagonal:", diagonal)
```

#### Ejercicio 10 : Dada una matriz 5x5 de enteros aleatorios entre 1 y 20, calcula la suma de cada fila y columna

```python
matriz = np.random.randint(1, 21, (5, 5))

suma_filas = matriz.sum(axis=1)
suma_columnas = matriz.sum(axis=0)

print("Matriz:\n", matriz)
print("Suma de filas:", suma_filas)
print("Suma de columnas:", suma_columnas)
```
