Al descargar el archivo Automobile.csv se obtiene un conjunto de datos que representa características de automóviles para la predicción de su consumo de combustible. Cada fila del conjunto de datos corresponde a un automóvil y contiene las siguientes columnas:

- `mpg`: Millas por galón (consumo de combustible)

- `cylinders`: Cilindros en el motor

- `displacement`: Desplazamiento del motor (en pulgadas cúbicas)

- `horsepower`: Potencia del motor (caballos de fuerza)

- `weight`: Peso del automóvil

- `acceleration`: Aceleración (0-60 mph en segundos)

- `model_year`: Año del modelo (ej. 70 representa 1970)

- `origin`: Origen del automóvil (1: Estados Unidos, 2: Europa, 3: Asia)



### Tareas a realizar:




##1.    Cargar el conjunto de datos en una estructura adecuada para su manipulación.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_automobile = pd.read_csv("Automobile.csv")
df_automobile.head()
```

```python
df_automobile.tail()
```

```python
print("\n--- Informacion general")
df_automobile.info()

print("\n--- Estadisticas descriptivas")
df_automobile.describe()
```

##2.   Crear un gráfico de dispersión para visualizar las relaciones entre los pares de características numéricas `mpg` y `weight`.

```python
sns.scatterplot(x='weight', y='mpg', data=df_automobile, color='#8a2be2'  )
plt.title('Relación entre Peso y Consumo de Combustible')
plt.xlabel('Peso del automóvil (weigth)')
plt.ylabel('Millas por galón (mpg)')

plt.show()
```

Este es un gráfico de dispersión (scatter plot) nos ilustra la relación entre el Peso y el Consumo de Combustible de un conjunto de vehículos.El gráfico revela una correlación inversa donde los vehículos más ligeros, con pesos entre 1500 y 2500 unidades, demuestran ser los más eficientes al superar frecuentemente las 30 mpg y alcanzar picos de hasta 45 mpg; sin embargo, este grupo presenta una alta dispersión vertical que sugiere la influencia de otros factores técnicos en su rendimiento. Por el contrario, al superar las 4000 unidades de peso, la eficiencia se ve drásticamente limitada, concentrándose la mayoría de los casos en un rendimiento bajo de entre 10 y 15 mpg. Esta tendencia indica que, mientras que en autos pequeños intervienen variables adicionales como la aerodinámica, en los vehículos pesados la masa se convierte en el factor determinante que condiciona inevitablemente un alto consumo de combustible.

##3.   Generar un histograma del consumo de combustible (`mpg`) para analizar su distribución.

```python
plt.figure(figsize=(8, 5))
plt.hist(df_automobile['mpg'], bins=20, color='#C8A2C8', edgecolor='black')
plt.xlabel('Consumo de combustible (mpg)')
plt.ylabel('Frecuencia')
plt.title('Distribución del consumo de combustible')
plt.show()
```

El histograma nos muestra la Distribución del consumo de combustible de un conjunto de automóviles, medido en millas por galón (mpg).Este revela una distribución de consumo con un marcado sesgo a la derecha, donde la mayor concentración de vehículos se agrupa en el rango de 13 a 20 mpg, alcanzando su punto máximo de frecuencia entre los 13 y 15 mpg. Aunque el rango total de eficiencia es amplio, extendiéndose desde los 9 hasta los 47 mpg(aproximadamente), la cantidad de automóviles disminuye notablemente a medida que aumenta el rendimiento, dejando a los modelos de alta eficiencia como casos excepcionales. A pesar de esta tendencia descendente, se observa un pequeño repunte en la frecuencia cerca de los 27 mpg, lo que rompe ligeramente la uniformidad de la caída antes de llegar a los valores más extremos de la muestra

##4.   Graficar un diagrama de caja (boxplot) de las características `cylinders` y `origin` para identificar posibles valores atípicos.

```python
df_automobile.boxplot(column="cylinders", by="origin", grid=True)
plt.title("Cilindradas por origen")
plt.xlabel('Origen del automóvil')
plt.ylabel('Número de cilindros')
plt.show()
```

El gráfico boxplot nos muestra la distribución del número de cilindros de los motores de automóviles, comparando las tres regiones de origen: Europa, Japón y EE. UU.
Estados Unidos presenta la mayor variabilidad con una clara inclinación hacia motores potentes de 6 y 8 cilindros, Europa y Japón muestran un comportamiento mucho más concentrado y homogéneo, centrando casi la totalidad de su producción en vehículos de 4 cilindros. Esta distinción subraya un contraste entre el mercado estadounidense, caracterizado por su diversidad y preferencia por la potencia, y los mercados europeo y japonés, enfocados en la estandarización y la eficiencia de motores más pequeños.

##5.   Realizar un gráfico de barras que muestre la distribución de automóviles según su año de modelo (`model_year`) y origen (`origin`).

```python
grouped = df_automobile.groupby(['model_year', 'origin']).size().unstack()
```

```python
grouped.plot(kind='bar', figsize=(12, 6), color=['#8a2be2', '#E74C3C', '#1E90FF'])
plt.xlabel('Año del modelo')
plt.ylabel('Cantidad de automóviles')
plt.title('Distribución de automóviles por año y origen')
plt.legend(title='Origen')
plt.show()
```

Este gráfico de barras nos  muestra la distribución de automóviles fabricados entre 1970 y 1982, clasificados según su año de modelo y su región de origen. Este nos permite ver que duurante la década de los 70, la industria automotriz estuvo marcada por un claro dominio de Estados Unidos, alcanzando su pico máximo en 1973; sin embargo, hacia finales del periodo la producción estadounidense mostró una tendencia a la baja, permitiendo que Japón experimentara un crecimiento constante hasta superar por primera vez a sus competidores en 1980. Ademas se demuestra como Europa se mantuvo como el productor con menor volumen, conservando una participación estable y sin grandes fluctuaciones a lo largo de todos los años representados.
